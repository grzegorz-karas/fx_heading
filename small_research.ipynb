```python
def read_close_price(contract_symbol):
    try:
        url = 'https://stooq.pl/q/d/l/?s=' + contract_symbol + '&i=d'
        dt = pd.read_csv(url,index_col=0)
        
        return dt['Zamkniecie']
    except:
        return None 
```

```python
months_letters = ('F','G','H','J','K','M','N','Q','U','V','X','Z')
months_numbers = np.arange(1,12)
```

```python
contracts = ['FEUR' + month_letter + '20' for month_letter in months_letters]
```

```python
contracts = 'FEURZ19' + contracts
```

```python
contracts_prices = {contract_symbol: read_close_price(contract_symbol) for contract_symbol in contracts}
```

```python
contracts_prices = pd.DataFrame(contracts_prices)
```

```python
contracts_prices.plot()
plt.show()
```
